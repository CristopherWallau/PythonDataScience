```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("harlfoxem/housesalesprediction")

print("Path to dataset files:", path)
```

```python
import pandas as pd
pd.set_option('display.max_columns', 21)
```

```python
arq = pd.read_csv('/kaggle/input/housesalesprediction/kc_house_data.csv')
```

```python
arq.shape
arq.info()
```

```python
arq.head()
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Define the features (independent variables) and target (dependent variable)
x = arq.drop(['id', 'date','zipcode','long','lat','price'], axis=1)  # Drop 'sqft_lot' and 'price' and use the rest as features
y = arq['price']     # Target can be a Series

# Split the data into training and testing sets (optional but good practice)
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.30, random_state=42)

print("Shape of x_treino:", x_treino.shape)
print("Shape of x_teste:", x_teste.shape)
print("Shape of y_treino:", y_treino.shape)
print("Shape of y_teste:", y_teste.shape)
```

```python
modelo = LinearRegression()
modelo.fit(x_treino, y_treino)
```

```python
modelo.score(x_teste,y_teste)
```
